From the given iris dataset, predict the optimum number of clusters and represent it visually.

 USING : K-Means Clustering

Importing necessary libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

Load the DATASET

```python
from sklearn.datasets import load_iris

dataset = load_iris()
dataset
```

Converting Dataset to a DataFrame

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)

df.head()
```

```python
# shape = (rows,columns)
df.shape
```

```python
# name of columns
df.columns
```

```python
# datatypes of columns
df.dtypes
```

```python
# statistical data of dataset
df.describe()
```

```python
# checking for null values

df.isnull().sum()
```

```python
X = dataset.data
Y = dataset.target
```

DATA VISUALIZATION

```python
# Scatter plot of sepal length and width

plt.scatter(X[Y == 0,0], X[Y == 0,1], label = "Setosa")
plt.scatter(X[Y == 1,0], X[Y == 1,1], label = "Versicolor")
plt.scatter(X[Y == 2,0], X[Y == 2,1], label = "Virginica")
plt.legend(loc = 'upper right')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# Scatter Plot of Petal length and width

plt.scatter(X[Y == 0,2], X[Y == 0,3], label = "Setosa")
plt.scatter(X[Y == 1,2], X[Y == 1,3], label = "Versicolor")
plt.scatter(X[Y == 2,2], X[Y == 2,3], label = "Virginica")
plt.legend(loc = 'lower right')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

APPLYING K-MEANS ALGORITHM TO FIND THE OPTIMUM NUMBER OF CLUSTERS

```python
from sklearn.cluster import KMeans

wcv = [] # WCV - Within Cluster Variation

for i in range(1,16):
    km = KMeans(n_clusters = i)
    km.fit(X)
    wcv.append(km.inertia_)

plt.plot(range(1,16), wcv)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCV')    
plt.show()
```

Creating the K-Means classifier

```python
km = KMeans(n_clusters = 3)
y_kmeans = km.fit_predict(X)
y_kmeans
```

```python
df['clusters'] = y_kmeans
df.head()
```

Visualizing thr clusters of sepal length and width

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0,1], label = 'Versicolor')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1,1], label = 'Setosa')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2,1], label = 'Virginica')

# Plotting Centroids

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], label = "Centroids")
plt.legend(loc = 'upper right')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

Visualizing the clusters of petal length and width

```python
plt.scatter(X[y_kmeans == 0, 2], X[y_kmeans == 0, 3], label = 'Versicolor')
plt.scatter(X[y_kmeans == 1, 2], X[y_kmeans == 1, 3], label = 'Setosa')
plt.scatter(X[y_kmeans == 2, 2], X[y_kmeans == 2, 3], label = 'Virginica')

# Plotting the centroids

plt.scatter(km.cluster_centers_[:, 2], km.cluster_centers_[:,3], label = 'Centroids')
plt.legend(loc='lower right') 
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

CONCLUSION :  We predicted the optimum number of clusters i.e. " 3 " and scatter plotted it
