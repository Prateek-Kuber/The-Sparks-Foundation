TASK 1: Prediction using Supervised ML

Predict the percentage of a student based on the number of study hours.

A simple Linear Regression task that involves two variables ie hours and score.

Importing necessary libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

Importing dataset and reading it

```python
scores_dataset = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
```

```python
scores_dataset.head(5)
```

```python
# shape = (rows , columns)
scores_dataset.shape
```

```python
# datatypes of columns
scores_dataset.dtypes
```

```python
# statistical details of the dataset
scores_dataset.describe()
```

```python
# checking for null values
scores_dataset.isnull().sum()
```

DATA VISUALIZATION

Plotting scores to time distribution

```python
scores_dataset.plot(x='Hours',y='Scores', style = 'o', markersize=5)
plt.title('Score to Time Distribution')
plt.xlabel('Hours of Study')
plt.ylabel('Score')
plt.grid()
plt.show()
```

The above graph represents that there is linear realtion between "HOURS OF STUDY" and "SCORE"

DATA PREPARATION


```python

X = scores_dataset.iloc[:, :-1].values  
Y = scores_dataset.iloc[:, 1].values

X.reshape(-1,1)
Y.reshape(-1,1)
```

Splitting Dataset into training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

Select and Train an ML Algorithm

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,Y_train)
```

Visualizing the Model

VISUALISING THE TRAINING DATASET

```python

# Visualising the Training dataset 
plt.scatter(X_train,Y_train)
plt.title('Training set')  
plt.plot(X_train,model.predict(X_train),color="green")
plt.xlabel('Hours of Study')  
plt.ylabel('Score') 
plt.grid()
plt.show()
```

```python
# Accuracy of the training set
model.score(X_train,Y_train)
```

VISUALISING THE TESTING DATA

```python
line = model.coef_*X + model.intercept_
plt.scatter(X, Y)
plt.title('Regression Line') 
plt.plot(X, line, color="green")
plt.xlabel('Hours of Study')  
plt.ylabel('Score') 
plt.grid()
plt.show()
```

MAKING PREDICTIONS

As we have trained our algorithm, let us make preditions

```python
Y_predicted = model.predict(X_test)
```

```python
compared_Dataset = pd.DataFrame({'Actual Score' : Y_test, 'Predicted score' : Y_predicted})
compared_Dataset
```

TESTING WITH OUR OWN DATA

According to the given task, what will be the score if a student studies 9.25 hours a day

```python
hours  = 9.25
own_prediction = model.predict([[hours]])
print(f"Predicted Score : {own_prediction[0]}")
```

EVALUATION OF THE MODEL

```python
model.score(X_test,Y_test)
```

```python
from sklearn import metrics

print("Mean Absolute Error : ",metrics.mean_absolute_error(Y_test,Y_predicted))
```
